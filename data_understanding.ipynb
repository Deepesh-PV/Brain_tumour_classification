In this section, we explore the dataset by visualizing sample images, inspecting class distributions, and verifying preprocessing steps. This helps ensure data quality and balance before model inference or training.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import datasets,transforms
```

```python
transformation=transforms.Compose([transforms.ToTensor()])
```

```python
train_data=datasets.ImageFolder("./brain_tumour/Tumour/train",transform=transformation)
```

```python
print(f"the classes in train data: {train_data.classes}")
class_idx=train_data.class_to_idx
idx_class={v:k for k,v in class_idx.items()}
print(f"The classes index: {class_idx}")
```

```python
counter=0
fig,axes=plt.subplots(1,4,figsize=(10,10))
for img,classes in train_data:
    if classes==0 and counter<4:
        axes[counter].imshow(img.permute(1,2,0))
        axes[counter].axis('off')
        axes[counter].set_title(f"class:glioma")
        counter=counter+1
plt.tight_layout()
plt.show()
```

```python
counter=0
fig,axes=plt.subplots(1,4,figsize=(10,10))
for img,classes in train_data:
    if classes==1 and counter<4:
        axes[counter].imshow(img.permute(1,2,0))
        axes[counter].axis('off')
        axes[counter].set_title(f"class:meningioma")
        counter=counter+1
plt.tight_layout()
plt.show()
```

```python
counter=0
fig,axes=plt.subplots(1,4,figsize=(10,10))
for img,classes in train_data:
    if classes==2 and counter<4:
        axes[counter].imshow(img.permute(1,2,0))
        axes[counter].axis('off')
        axes[counter].set_title(f"class:no_tumour")
        counter=counter+1
plt.tight_layout()
plt.show()
```

```python
counter=0
fig,axes=plt.subplots(1,4,figsize=(10,10))
for img,classes in train_data:
    if classes==3 and counter<4:
        axes[counter].imshow(img.permute(1,2,0))
        axes[counter].axis('off')
        axes[counter].set_title(f"class:pituatory")
        counter=counter+1
plt.tight_layout()
plt.show()
```

```python
class_counts={0:0,1:0,2:0,3:0}
```

```python
for img,classes in train_data:
    class_counts[classes]=1+class_counts[classes]
```

```python
print(class_counts)
```

```python
sns.barplot(x=list(class_idx.keys()),y=list(class_counts.values()))
plt.title("Number of Images per Class")
plt.ylabel("Image Count")
plt.xlabel("Tumor Type")
plt.xticks(rotation=45)
plt.show()
```
